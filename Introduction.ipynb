## Population and Samples
* During research always ask what is the population of interest and what is the sample,
* Sampling means select portion of larger population and study that portion to gain info about the population. Data are the result of sampling from population.
* Observing entire population is expensive, so sampling is needed.
* From sample data we calculate statistics. Which is estimate of population parameter. A parameter is numerical characteristics of the whole population that can be estimated by statistics.
* One main concern is how well statistics estimates parameter. Which is depends on how well the sample represents population. 
* Sample should have same characteristics as population it is representing.
* Sampling error = parameter - statistics
    - Difference between metrics of population and metrics of sample.

* Before collecting sample define target population
    - Who we are measuring, african american male, older american male
    
### Random sampling
* Each member of population has equal chance of being selected.
* True random sampling is done by replacement. Surveys are done without replacement. Because population is very large compared to sample, sampling without replacement or with replacement is almost same.

#### Simple random sampling
* Each case in population has equal chance of being selected. Ex. Drawing name from well mixed slips from bowl
* Knowing that one case is included in the case does not provide useful information about what other case are included.
* Not good when sample size is small.
```
Series.sample(50, random_state = 1)
```
![simple random sampling](images/simple_random_sampling.JPG)

#### Stratified sampling
* Population is divided into group called strata.Strata are chosen so similar case are grouped together. Simple random sampling is employed in each stratum. Ex. group the cricket team of all nations and draw 4 names from each group. Useful when we want same number of cases from individual groups.
```
for name, grp in df.groupby('col_name')
    # sampling within each group 
```
* Choose strata effectively, minimize variability within stratum, Either make more granular stratum, And maximize variability between the stratum. Stratification criteria must be strongly correlated with property we try to measure.
![stratified_sampling](images/stratified_sampling.JPG)

#### Cluster sample
* We break up population in many group called clusters. We sample fixed number of clusters and includes all the individuals from sampled clusters.
* Useful when lots of case to case variability within the cluster but 2 clusters do not look different from one another.
* It incur sampling variance, because observation in same cluster have similar values on variables of interest, do not measure unique information.
![cluster sampling](images/cluster_sampling.JPG)

#### Multistage sample:
* Rather than keeping all cases from clusters, we collect random sample within each selected clusters.
![Multistage sampling](images/multistage_sampling.JPG)

#### Systematic sample
* Randomly select starting point and take every nth data from listing of the population. Ex. Phone book has 20000 people listing. To choose 400 people from it, we can randomly choose 1 entry then choose every 15th name.

* Using technique called proportionate allocation we can make sure our sample is representative. In college we have 55% female and 45% male, to select 1000 sample we can select 550 female and 450 male randomly.
* Using stratified sampling we can reduce the variance, we can get rid of inter stratum variance compared to Simple random sampling. So, when in doubt stratify. It reduce selection bias, but make sure it does not trigger non response bias.

* Look at the data understand from where it came from. Read documentation and what type of sampling used.
* When we have randomness, our estimate of mean, proportion and total are unbiased, meaning our estimate are equal to the population value on average.

### Sampling bias
* Sample which is not representative of population is biased.
* It occurs when some members is less likely to be chosen as others.
* When person is taking sample, it is possible that he chooses case as per his interest. This will create bias.
* Convenience sampling: Individual who are accessible are more likely to be chosen in sample.
* Self selected/ voluntary response samples:  Responses by the people who are willing to response, call in survey.
* Collecting data or asking questions in a way that influences response.
* Non response: People with strong positive or negative opinion answer the survey.
* In non response bias there is a random sample which was survey, but some of them dont want to answer, in voluntary response bias there is no random sample.

### Inference from Biased data

#### Quasi randomization Pseudo Randomization Approach
* Combine biased sample with true randomized sample. Stack 2 dataset together. Add column which indicate which observation is from where, like 1 for biased and 0 for random
* Fit logistic regression model. Predict probability of having 1 in newly added column (Probability of being in biased sample) with other variable.
* Invert predicted probability for biased sample, treat as survey weight in standard survey analysis. 

#### Population  modeling
* Use predicted modeling to predict aggregated sample quantities on key variable of interest for population units not included in biased sample. 
* If we have good auxiliary info for everyone in population and have some extra data for part of population we can predict extra data with predictive model.

### Sampling biased population demonstration

````python
mean_weight_uni_students = 155
sd_weight_uni_student = 5
mean_weight_gym_goers = 185
sd_weight_gym_stuents = 5
gymperc = 0.30
totalPopSize = 40000
````

##### Create populaiton subgroups

````python
uni_students = np.random.normal(mean_weight_uni_students, sd_weight_uni_student, int(totalPopSize * (1-gymperc)))
students_at_gym = np.random.normal(mean_weight_gym_goers, sd_weight_gym_stuents, int(totalPopSize * gymperc))
````

##### Combine population to get total population

````python
population = np.append(uni_students, students_at_gym)
````

##### Plot entire university students

````python
plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
sns.distplot(uni_students).set_title("University student Only")
plt.xlim([140,200]);

### Plot gym goers

plt.subplot(3,1,2)
sns.distplot(students_at_gym).set_title("Gym goers only")
plt.xlim([140,200]);

### Plot both
plt.subplot(3,1,3)
sns.distplot(population). set_title("Full population of university")
plt.axvline(x=np.mean(population))
plt.xlim([140, 200]);
````

##### Sample from entire population

````python
numOfSamples = 5000
sampSize = 50

mean_distribution = np.empty(numOfSamples)

for i in range(numOfSamples):
    random_students = np.random.choice(population, sampSize)
    mean_distribution[i] = np.mean(random_students)
````

````python
plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
sns.distplot(population).set_title("Entire population")
plt.axvline(x=np.mean(population))
plt.xlim([140,200])

plt.subplot(2,1,2)
sns.distplot(mean_distribution).set_title("Sampling distribution of mean")
plt.axvline(x=np.mean(population))
plt.axvline(x=np.mean(mean_distribution), color="black")
plt.xlim([140,200]);
````

##### Sample from non representative sample (Convenient sample)

````python
mean_distribution_gym_goers = np.empty(numOfSamples)

for i in range(numOfSamples):
    gym_goer_sample = np.random.choice(students_at_gym, sampSize)
    mean_distribution_gym_goers[i] = gym_goer_sample.mean()
````

````python
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
sns.distplot(population).set_title("Entire population")
plt.axvline(x=population.mean())
plt.xlim([140,200])


plt.subplot(2,1,2)
sns.distplot(mean_distribution_gym_goers).set_title("Sampling distribution of mean of gym goers")
plt.axvline(x=np.mean(population))
plt.axvline(x=np.mean(mean_distribution_gym_goers), color='black')
plt.xlim([140,200]);
````

* When one variable changes other first variable is called **explanatory (Independent) variable**. Affected variable is called **Response (dependent) variable**. In randomized experience researcher controls explanatory variable and measure changes in response variable. Different value of explanatory variable is called treatment.
* Variable which can affect study are called **Lurking(confounding) variable**. Extraneous variable that affect both the explanatory and response variable and make it seem like there is relationship between them. To prove that the explanatory variable causes the change in response variable we need to get reed of such lurking variable.
* This can achieve by random assignment of experimental units to different treatment groups. Doing so will spread equally all potential lurking variables.

* **Control group** Does not receive treatment/drug/medications. It provides reference point to which we can measure impact of treatment group.
* When researcher keep the participants uninformed regarding in which group they are, such study is called blind.

## Observational vs Experimental study
### Observational
* Collect the data in a way that does not directly interfere how the data arise.Ex. Collect info via survey, review medical or company record.
* Researcher only observe the data which arise. Such study can provide naturally occurring association between variables, but can not show causal connection.
* To make causal statement in observational study just get rid of all confounding variables, but it is not possible.
* It can be prospective study: Data is collected during the study or retrospective study:using past information.

### Experimental
* Individual are assigned in treatment and control group. If individual are assigned randomly such experiment is called randomized experiment. Generally control group receives placebo (fake treatment). Treatment group receives actual treatment which needs to be evaluated.
* Association does not imply causation, causation can only be inferred by randomized experiment.
* Randomized experiment build on 4 principles
1. Controlling: Researchers tries best to control differences between treatment and control groups. Ex. To control the water consumption effect every individual will take a pill with 12 oz of water.
2. Randomization: Researcher randomize patients in groups for account for variables which can not be controlled. 
3. Replication: More case researcher observe, more sure we are for effect of explanatory variable on response variable. We can get larger sample and replicate same experiment or other scientist replicate same experiment on same sample.
4. Blocking: Some variables can influence the response. First group individual based on this variables into blocks and then randomize cases within each block to different groups.Ex. Divide patient in low risk and high risk blocks. Randomly assign high patient from each group to control and treatment group. Which ensure both group has equal number of low and high risk patients.
![blocking](images/blocking.JPG)
* Explanatory variables are conditions we can impose on experimental units, blocking variables are characteristics that the experimental units come with, that we would like to control for.

### Random sampling vs Random assignment
* Random sampling occurs when subjects are selected from population. If subject are selected randomly then each subject in population equally likely to be selected and sample is representative of population. Result can be generalizable to the population.
* Random assignment occurs in experimental study. Subjects has different characteristics, to divide equally we randomly assign them in treatment and control group.

## Intro to inference
* 48 male bank supervisor given the same personal file, asked to judge whether the person should promoted or not. Files are identical except gender of applicant. Random assignment.
![inference](images/inference.JPG)
* 2 competing claim
1. NULL HYPOTHESIS: Nothing going on, promotion and gender is independent. Observed difference simply due to chance.
2. ALTERNATIVE HYPOTHESIS: there is a gender discrimination. Observed difference is NOT due to chance.
![null_alternative](images/null_alternative.JPG)
- We start with null hypothesis which is status quo.
- Set alternative hypothesis which is a research question i.e what we are testing for
- Conduct hypothesis test under the assumption that null hypothesis is true, either via simulation or theoretical methods.
    - If result do not provide convincing evidence stick with null hypothesis.
    - Otherwise reject null hypothesis in favor of alternative.
* Simulation method.
    - Take total 48 card. 13 face card and 35 number card.
    - Face card means NOT promoted
    - Non face card is promoted
    - Shuffle it and divide in 2 groups male and female.
    - Count number of card in each group that is promoted, not promoted.
    - Note the difference in 2 groups.
    - Repeat steps many times.
    - Build a distribution plot.
![discrimination_plot](images/discrimination_plot.JPG)

* If result of simulation look like a data, the difference between proportion is just a chance. Otherwise there is an actual discrimination.
* Evaluate probability of observing an outcome at least as extreme as the observed in the original data (p-value). If this probability is low reject null hypothesis in favor of alternative.
